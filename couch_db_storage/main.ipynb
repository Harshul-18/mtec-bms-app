```python
import couchdb
import pandas as pd

def retrieve_all_documents(username, password, db_name, save=False):
    couchdb_url = f"http://{username}:{password}@localhost:5984/"
    couch = couchdb.Server(couchdb_url)
    db = couch[db_name]
    docs = []
    for doc_id in db:
        doc = db[doc_id]
        docs.append(doc)
    data = pd.DataFrame(docs)
    if save==True:
        data.to_csv(f'{db_name}_data.csv')
    return data

if __name__ == "__main__":
    data = retrieve_all_documents(username='admin',
                                password='root',
                                db_name='vehicles',
                                save=True)

data['timestamp'] = pd.to_datetime(data['timestamp'], errors='coerce')
data = data.dropna(subset=['timestamp'])
data['timestamp'] = data['timestamp'].dt.tz_localize(None)
```

```python
import pandas as pd
import couchdb

def retrieve_data(username, password, db_names, save=False):
    couchdb_url = f"http://{username}:{password}@localhost:5984/"
    couch = couchdb.Server(couchdb_url)
    all_docs = []

    for db_name in db_names:
        db_name = '_'.join(db_name.lower().split())
        if db_name in couch:
            db = couch[db_name]
            for doc_id in db:
                doc = db[doc_id]
                all_docs.append(doc)
        else:
            print(f"Database '{db_name}' does not exist.")

    # Convert list of documents to DataFrame
    data = pd.DataFrame(all_docs)
    if save:
        data.to_csv(f'all_models_data.csv', index=False)
    return data

# Example usage
db_names = ['Model 1']
df = retrieve_data(username='admin', password='root', db_names=db_names, save=True)
df.model.unique()
```
